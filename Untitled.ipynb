```python
#step 1
#Import all the required libraries besfore starting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('netflix_titles.csv')
```

```python
#Data cleaning 
#drop rows where director is missing 
df.dropna(subset=['director'],inplace=True)
```

```python
df.isnull().sum()
```

```python
#fill missing values in 'country' and 'rating'columns
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
#convert dateadded to datetime
#first removes any unwanted spaces before or after the date strings with (str.strip()).
df['date_added'] = df['date_added'].str.strip() 
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df['date_added'].head(5)
```

```python
#create a new column 'year added'
df['year_added'] = df['date_added'].dt.year
```

```python
df.columns
```

```python
#step 6 EDA 
#task 1: count of movies vs tv shows 
sns.countplot(data=df,x='type')
plt.title('Count of movies vs TV shows on Netflix')
plt.show()
```

```python
#task 2: top 10 countries with most titles
top_countries=df['country'].value_counts().head(10)
top_countries.plot(kind='barh',color='skyblue')
plt.title('TOp 10 countries on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

```python
#task 3: Number of titles over the years
df['year_added'].value_counts().sort_index().plot(kind='bar',color='blue')
plt.title('Number of Titles added over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
#task 4: MOST COMMON GENRES
#combine all genres into a single list 
from collections import Counter
all_genres=','.join(df['listed_in']).split(',')
genre_count=Counter(all_genres)
most_common_genres=genre_count.most_common(10)

#convert to dataframe
genres_df=pd.DataFrame(most_common_genres,columns=['Genre','Count'])

#plot top 10 Genres
sns.barplot(data=genres_df, y='Genre', x='Count', color='coral')
plt.title('Top 10 Most Common Genres')
plt.show()
```
